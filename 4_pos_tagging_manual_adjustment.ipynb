```python
# Precons
INPUT_FILE=os.path.join(OUT_DIR,'..','pos_log','examples_tagged_upos.csv')
INPUT_FILE=os.path.join(OUT_DIR,'..','pos_log','examples_tagged_upos_majority.csv')
# if file does not exist, throw an error
if not os.path.exists(INPUT_FILE):
    precon='2_pos_tagging_create.ipynb'
    raise FileNotFoundError(f"File '{INPUT_FILE}' not found. Run '{precon}' first.")

#OUTPUT_FILE=os.path.join(OUT_DIR, 'example_event_templates_3_manual_tagged.csv')
OUTPUT_FILE=os.path.join(OUT_DIR,'..','pos_log','ground_truth_wip_upos.csv')
```

```python
import pandas as pd
tagged_examples:pd.DataFrame

if os.path.exists(OUTPUT_FILE):
    print(f"File '{OUTPUT_FILE}' already exists. Go on editing or remove it.")
    tagged_examples=pd.read_csv(OUTPUT_FILE)
else: 
    # columns: Dataset,Index,Example,EventId,EventTemplate, Example_Splitted, MaskedExampleSplitted, [tagger], TagComparison
    tagged_examples=pd.read_csv(INPUT_FILE)

tagged_examples.head()
```

```python
from nlp.pos import TagComparison

log_lines=tagged_examples['Example'].to_list()
log_lines_splitted=[eval(x) for x in tagged_examples['Tokens'].to_list()]
tag_comparisons:list[TagComparison]=[eval(x) for x in tagged_examples['TagComparison'].to_list()]
majorities:list[list[str]]=[eval(x) for x in tagged_examples['Majority'].to_list()]
```

```python
manually_tagged=[]
if 'ManualTagging' in tagged_examples.columns:
    manually_tagged=[eval(x) for x in tagged_examples['ManualTagging'].to_list()]
else:
    # This now keeps the 80% and 60% majorities
    manually_tagged=[m.copy() for m in majorities]
print(manually_tagged)
```

# GUI Logs

```python
from gui_and_helper import GuiLogs

FILTER_DONE=False

gui_logs:list[GuiLogs]=[]
for i, (log_line, log_line_splitted, tag_comparison, majorities_per_line, manual_tags) in enumerate(zip(log_lines, log_lines_splitted, tag_comparisons, majorities, manually_tagged)):
    if not FILTER_DONE or None in manual_tags:
        gui_logs.append(GuiLogs(i, log_line, log_line_splitted, tag_comparison, majorities_per_line, manual_tags))

gui_logs[:10]
```
